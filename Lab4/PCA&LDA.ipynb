## 1. Apply the PCA and LDA techniques on the IRIS dataset from sklearn - are both applicable? ##

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis 
from sklearn.datasets import load_iris 
import matplotlib.pyplot as plt 
data = load_iris() 
X = data.data 
y = data.target 
target_names = data.target_names 
lda = LinearDiscriminantAnalysis(n_components=2) 
X_lda = lda.fit(X, y).transform(X) 
plt.figure() 
colors = ['navy', 'turquoise', 'darkorange'] 
lw = 2 

for color, i, target_name in zip(colors, [0, 1, 2], target_names): 
    plt.scatter(X_lda[y == i, 0], X_lda[y == i, 1], alpha=.8, color=color, label=target_name) 
 
plt.legend(loc='best', shadow=False, scatterpoints=1) 
plt.title('LDA of IRIS dataset') 
plt.show()
```

```python
from sklearn.decomposition import PCA 
pca=PCA(n_components=3) 
X_new = pca.fit_transform(X) 
plt.figure() 
colors = ['navy', 'turquoise', 'darkorange'] 
lw = 2 

for color, i, target_name in zip(colors, [0, 1, 2], target_names): 
    plt.scatter(X_new[y == i, 0], X_new[y == i, 1], alpha=.8, color=color, label=target_name) 
    
plt.legend(loc='best', shadow=False, scatterpoints=1) 
plt.title('PCA of IRIS dataset') 
plt.show() 
```

## 2. Explain the difference between two approaches. ##
**PCA** , It's primarily a technique for dimensionality reduction while preserving as much variance in the data as possible. It doesn't take into account the class label or any information about class separation. The output of PCA is set of orthogonal components (Principal Component) that capture the maximum variance in data. PCA assumes that the data is centered (mean subtracted) and that the principal components are the linear combination of the original features. 
 **LDA** , is a supervised dimensionality reduction technique which considers both variance of the data and the difference between classes. The output of LDA is set of linear discriminants that maximizes separation between classes. LDA assumes that the data is normally distributed within each class and that the classes have identical covariance matrices.es.ures.

PCA is a preferred technique when there is a need to reduce the number of features in a dataset while preserving as much variance as possible. It excels in visualizing high-dimensional data in 2D or 3D, which helps in uncovering underlying patterns or clusters. Since PCA does not consider class labels, it is best suited for unsupervised tasks such as clustering or when there is an interest in understanding the overall structure of the data without focusing on specific classes.

In contrast, LDA is most suitable for classification tasks where there is a need to reduce dimensionality while also maximizing the separation between different classes. LDA is a supervised learning technique that leverages class labels to determine the linear combinations of features that most effectively distinguish between classes. LDA assumes that the data within each class follows a normal distribution and that the classes share similar covariance matrices. Consequently, LDA performs optimally when these assumptions are satisfied.
